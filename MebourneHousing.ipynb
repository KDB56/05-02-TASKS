```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
maldf=pd.read_csv('Melbourne_housing.csv')
maldf
```

```python
maldf.nunique()
```

```python
maldf=maldf.drop(columns=['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'])
```

```python
maldf
```

```python
maldf=maldf.dropna()
maldf.isnull().sum()
```

```python
maldf.shape
```

```python
maldf['Suburb'].unique()
```

```python
print(maldf['Suburb'].value_counts())
```

```python
maldf['Type'].unique()
```

```python
maldf['SellerG'].unique()
```

```python
maldf['Regionname'].unique()
```

```python
maldf['CouncilArea'].unique()
```

```python
maldf['Method'].unique()
```

```python
maldf['Type']=maldf['Type'].replace({'h':0,'u':1,'t':2})
maldf['Regionname']=maldf['Regionname'].replace({'Northern Metropolitan':0,'Western Metropolitan':1,'Southern Metropolitan':2,'Eastern Metropolitan':3,'South-Eastern Metropolitan':4,'Northern Victoria':5,'Eastern Victoria':6,'Western Victoria':7})
maldf['Method']=maldf['Method'].replace({'S':0,'SP':1,'VB':2,'PI':3,'SA':4})
maldf
```

```python
maldf.info()
```

```python
maldf=pd.get_dummies(maldf,dtype=int)
maldf
```

```python
x=maldf.drop(columns=['Price'])
y=maldf['Price']
y

```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
LinModel=LinearRegression()
LinModel.fit(x_train,y_train)
```

```python
LinModel.predict(x_test)
```

```python
LinModel.score(x_test,y_test)
```

```python
LinModel.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso
reg2= Lasso(alpha=25,max_iter=100,tol=0.1)
reg2.fit(x_train,y_train)
```

```python
reg2.predict(x_test)
```

```python
reg2.score(x_test,y_test)
```

```python
reg2.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
reg3= Ridge(alpha=25,max_iter=100,tol=0.1)
reg3.fit(x_train,y_train)
```

```python
reg3.predict(x_test)
```

```python
reg3.score(x_test,y_test)
```

```python
reg3.score(x_train,y_train)
```
